```python
credits=pd.read_csv('tmdb_5000_credits.csv')
movie=pd.read_csv('tmdb_5000_movies.csv')
```

```python
#merge the both dataset based on the common feature
movies=movie.merge(credits,on="title")
```

```python
#then select the columns based on the our domain knowledge to recommend the movies
#genres
#keywords
#overview
#title
#id
#cast
#crew

movies = movies[["genres","id","keywords","overview","title","cast","crew"]]
```

```python
#literal eval:as the name suggest it evaluate the string to give give the correct datatype value
import ast
ast.literal_eval("[5,8,9]")
```

```python
#now i combine the geners,keywards,cast and crew into single feature to make efficient 
def get_genres(a):
    l=[]
    for i in ast.literal_eval(a):
        l.append(i['name'])
    return l
```

```python
genres=movies["genres"].apply(get_genres)
```

```python
movies["genres"]=genres
```

```python
movies
```

```python
key=movies["keywords"].apply(get_genres)
```

```python
movies["keywords"]=key
```

```python
movies
```

```python
def get_cast(z):
    counter=1
    l=[]
    for i in ast.literal_eval(z):
        if(counter<=3):
            l.append(i['name'])
            counter=counter+1
        else:
            break
    return l
```

```python
cast=movies["cast"].apply(get_cast)
```

```python
movies["cast"]=cast
```

```python
movies
```

```python
ast.literal_eval(movies.crew[0])
```

```python
def get_crew(a):
    l=[]
    for i in ast.literal_eval(a):
        for j in i.values():
            if(j=="Director"):
                l.append(i["name"])
    return l
            
```

```python
director=movies["crew"].apply(get_crew)
```

```python
movies["crew"]=director
```

```python
movies.head(10)
```

```python
movies["overview"] = movies["overview"].apply(lambda x : x.split())
```

```python
movies.head(10)
```

```python
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["overview"]=movies["overview"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies["tags"]=movies["genres"] + movies["keywords"] + movies["overview"] + movies["cast"] + movies["crew"]
```

```python
movies.head(1)
```

```python
movies = movies[["id","title","tags"]]
```

```python
movies.head(10)
```

```python
movies["tags"] = movies["tags"].apply(lambda x : " ".join(x))
```

```python
movies.head(10)
```

```python
movies["tags"] = movies["tags"].apply(lambda x : x.lower())
```

```python
movies.head(10)
```

```python
movies.shape
```

```python
#convert the tages into vectors then based on the distance we recommand the movies
#some of the words are repeating different types like actor,actors so we convert the text to vector these two word are contributed in seperately to avoid these we are using the stem technique to convert actor and actors into same common words like actor
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem("actors")
```

```python
def get_common(text):
    final=[]
    for i in text.split():
        final.append(ps.stem(i))
        
    return " ".join(final)
        
```

```python
movies["tags"] = movies["tags"].apply(get_common)
```

```python
movies.head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(movies["tags"]).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cs = cosine_similarity(vectors)
ls=pd.DataFrame(cs)
```

```python
cs[0]
```

```python
def recommand(movie):
    mov_index = movies[movies["title"]==movie].index[0]
    movie_list = sorted(enumerate(cs[movies[movies["title"]==movie].index[0]]),reverse=True,key=lambda x:x[1])[:6]
    for i in movie_list:
        print(movies["title"].iloc[i[0]])
```

```python
recommand("Harry Potter and the Half-Blood Prince")
```
